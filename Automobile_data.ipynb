```python
# get_dummies Encoding
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df1
```

```python
# df1 has encoded datas. But we also need other datas so, we perform concatenation.
df2=pd.concat([df,df1],axis=1)
df2
```

```python
# drop unwanted columns
df2.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Fuel_Type_Electric'],axis=1,inplace=True)
df2
```

```python
# To Replace string in data of numerical columns i.e, km/kg and km/l in mileage
df2['Mileage']=df2['Mileage'].str.replace('km/kg','')
df2['Mileage']=df2['Mileage'].str.replace('kmpl','')
df2['Engine']=df2['Engine'].str.replace('CC','')
df2['Power']=df2['Power'].str.replace('bhp','')
# In column mileage, null data is present which is a string as 'null'
# So we have to replace the data with '0'
df2['Mileage']=df2['Mileage'].str.replace('null','0')   # '0' is passed as string.
df2['Power']=df2['Power'].str.replace('null','0')       # Here, datatype is not changed from object as int/float. In next step dtype will change.
df2
```

```python
# Changing datatypes
df2['Engine']=df2['Engine'].astype(float)
df2['Mileage']=df2['Mileage'].astype(float)
df2['Power']=df2['Power'].astype(float)
df2.dtypes
```

```python
df2.isna().sum()
```

```python
# We have replaced 'null' in engine and power with '0'. But mileage and power values will never be 0. So, it will be treated as missing values
import numpy as np
df2.loc[df2.Mileage==0,'Mileage']=np.NaN
df2.loc[df2.Power==0,'Power']=np.NaN
df2.isna().sum()
```

```python
# Fill missing values by mean,median or mode. Seat capacity will be integer number so use mode
df2['Mileage']=df2['Mileage'].fillna(df2['Mileage'].mean())
df2['Engine']=df2['Engine'].fillna(df2['Engine'].mean())
df2['Power']=df2['Power'].fillna(df2['Power'].mean())
df2['Seats']=df2['Seats'].fillna(df2['Seats'].mode()[0])
df2
```

```python
df2.isna().sum()
```
